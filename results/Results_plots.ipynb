```python
model4a_predictions = pd.read_csv("/Users/rickfontenot/Dropbox/SMU/Capstone/Github/HNN-Capstone/Results/model4a_predictions.csv")
model4a_predictions.head()
```

```python
model4a_predictions.fine_label.unique()
```

```python
model4a_predictions.sort_values(by=['fine_label'], inplace=True)
```

```python
model4a_predictions.fine_label.unique()
```

```python
from sklearn.metrics import confusion_matrix

def plot_confusion(cf_matrix):
    plt.figure(figsize=(20, 16))
    sns.set(font_scale=2)
    group_names = ['True Neg','False Pos','False Neg','True Pos']
    group_counts = ['{0:0.0f}'.format(value) for value in
    cf_matrix.flatten()]
    group_percentages = ['{0:.2%}'.format(value) for value in
    cf_matrix.flatten()/np.sum(cf_matrix)]
    labels = [f'{v1}\n{v2}\n' for v1, v2 in
    zip(group_counts,group_percentages)]
    labels = np.asarray(labels).reshape(10,10)
    x_axis_labels = ['Ankle boot', 'Bag', 'Coat', 'Dress', 'Pullover', 'Sandal',
       'Shirt', 'Sneaker', 'T-shirt/top', 'Trouser']
    y_axis_labels = ['Ankle boot', 'Bag', 'Coat', 'Dress', 'Pullover', 'Sandal',
       'Shirt', 'Sneaker', 'T-shirt/top', 'Trouser']
    ax = plt.axes()
    sns.heatmap(cf_matrix, annot=labels, fmt='', cmap='Blues',xticklabels=x_axis_labels, yticklabels=y_axis_labels)
    #sns.heatmap(cf_matrix, annot=labels, fmt='', cmap='Blues')
    ax.set_title('             Model 4A Confusion Matrix\n')
    plt.xlabel("Predicted Category")
    plt.ylabel("True Category") 


cm = confusion_matrix(model4a_predictions.fine_label,model4a_predictions.fine_label_prediction) 
plot_confusion(cm)
```
